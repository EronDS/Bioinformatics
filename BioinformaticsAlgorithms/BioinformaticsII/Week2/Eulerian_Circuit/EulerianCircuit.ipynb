```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python

class EulerianCircuit:
    def __init__(self):
        self.graph = {}
    def dict_from_file(self,file):
        with open(file, 'r') as file:
            self.graph = dict((line.strip().split(' -> ') for line in file))
            for key in self.graph:
                self.graph[key] = self.graph[key].split(',')
        return self.graph
    
    def find_eulerian_circuit(self):
        g = nx.DiGraph(self.graph)
        plt.figure(figsize=(20,10))
        nx.draw_networkx(g, with_labels = False,node_color = 'salmon',font_color='white')
        if nx.is_eulerian(g):
            eulerian_circuit = list(nx.algorithms.euler.eulerian_circuit(g))
        else:
            print('Not an Eulerian Graph')
            return
        
        eulerian_ = [] 
        for i in eulerian_circuit:
            if i == eulerian_circuit[0]:
                eulerian_.append(i[0])
                eulerian_.append(i[1])
            else:
                eulerian_.append(i[1])
        
        f = open('eulerian_circuit.txt','w')
        f.write('->'.join(str(i) for i in eulerian_))
        f.close()
        return
```

```python
eu = EulerianCircuit()
eu.dict_from_file('/content/dataset_203_2.txt')
eu.find_eulerian_circuit()
```
