```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python

class EulerianPath:
    def __init__(self):
        self.graph = {}
    def dict_from_file(self,file):
        with open(file, 'r') as file:
            self.graph = dict((line.strip().split(' -> ') for line in file))
            for key in self.graph:
                self.graph[key] = self.graph[key].split(',')
        return self.graph
    
    def find_eulerian_path(self):
        g = nx.DiGraph(self.graph)
        plt.figure(figsize=(20,10))
        nx.draw_networkx(g, with_labels = True,node_color = 'skyblue',font_color='darkblue')
        eulerian_path = list(nx.algorithms.euler.eulerian_path(g))
        
        eulerian_ = [] 
        for i in eulerian_path:
            if i == eulerian_path[0]:
                eulerian_.append(i[0])
                eulerian_.append(i[1])
            else:
                eulerian_.append(i[1])
        
        f = open('eulerian_path.txt','w')
        f.write('->'.join(str(i) for i in eulerian_))
        f.close()
        return
```

```python
eu = EulerianPath()
eu.dict_from_file('/content/dataset_203_6.txt')
eu.find_eulerian_path()
```
